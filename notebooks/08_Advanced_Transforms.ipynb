```python
import numpy as np, matplotlib.pyplot as plt
from numpy.fft import fft2, fftshift
from skimage import data
img = data.camera(); spec = np.log1p(np.abs(fftshift(fft2(img))))
plt.figure(figsize=(8,4));
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Image'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(spec, cmap='inferno'); plt.title('Power Spectrum'); plt.axis('off')
```
