```python
import numpy as np, matplotlib.pyplot as plt
from skimage import filters, segmentation, data
img = data.camera(); thresh = filters.threshold_otsu(img); mask = img > thresh
plt.figure(figsize=(8,4));
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Image'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(mask, cmap='gray'); plt.title('Otsu Mask'); plt.axis('off')
```
