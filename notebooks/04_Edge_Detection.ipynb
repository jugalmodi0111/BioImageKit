```python
import numpy as np, matplotlib.pyplot as plt
from skimage import filters, feature, data
img = data.camera(); edges = feature.canny(img/255.0, sigma=1.5)
plt.figure(figsize=(8,4));
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Image'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(edges, cmap='gray'); plt.title('Canny'); plt.axis('off')
```
