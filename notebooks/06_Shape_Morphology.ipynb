```python
import numpy as np, matplotlib.pyplot as plt
from skimage import morphology, data
img = data.binary_blobs(length=128, blob_size_fraction=0.1, n_dim=2).astype('uint8')
opened = morphology.opening(img, morphology.disk(2))
plt.figure(figsize=(8,4));
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Binary'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(opened, cmap='gray'); plt.title('Opening'); plt.axis('off')
```
