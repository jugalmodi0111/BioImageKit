```python
import numpy as np, matplotlib.pyplot as plt
from skimage import feature, data
img = data.camera(); lbp = feature.local_binary_pattern(img, 8, 1, method='uniform')
plt.figure(figsize=(8,4));
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Image'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(lbp, cmap='magma'); plt.title('LBP'); plt.axis('off')
```
