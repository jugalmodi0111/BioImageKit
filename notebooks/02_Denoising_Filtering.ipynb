```python
import numpy as np, matplotlib.pyplot as plt
from skimage import restoration, filters, data
img = data.camera(); noisy = img.astype(float) + np.random.normal(0, 10, img.shape)
noisy = np.clip(noisy, 0, 255).astype('uint8')
plt.figure(figsize=(8,4));
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Original'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(noisy, cmap='gray'); plt.title('Noisy'); plt.axis('off')
```
